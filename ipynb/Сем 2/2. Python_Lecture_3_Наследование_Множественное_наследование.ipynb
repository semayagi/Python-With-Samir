## Наследование



**Наследование** в объектно-ориентированном программировании создаёт отношение **"является"** между классами. Это ключевой принцип, позволяющий одному классу (**производному классу**, также известному как дочерний класс или подкласс) наследовать атрибуты и методы другого класса (**базового класса**, также называемого родительским классом или суперклассом).

#### Преимущества наследования:
- **Реализация общих методов** на уровне базового класса, чтобы избежать дублирования кода.
- Возможность **расширения или модификации** функциональности базового класса в производных классах для выполнения более специфичных задач.

### Визуализация отношений с помощью UML диаграмм

Для визуализации и понимания отношений между классами используются **диаграммы UML** (Унифицированный язык моделирования). Эти диаграммы представляют собой мощный инструмент для документирования архитектуры программного обеспечения.

#### Особенности UML диаграмм в контексте наследования:
- Классы изображаются в виде прямоугольных блоков с именем класса, размещённым в верхней части блока.
- Атрибуты и методы класса могут быть также указаны в блоке для детализации.
- Отношения наследования отображаются стрелками с полым треугольником на конце, направленными от производного класса к базовому.

Эти элементы диаграммы помогают разработчикам создавать чёткую и эффективную иерархию классов, минимизируя ошибки в архитектуре и облегчая понимание взаимосвязей внутри системы.

### Принцип подстановки Лисков (LSP)

**Принцип подстановки Лисков (LSP)** является фундаментальным понятием в объектно-ориентированном программировании, обеспечивающим правильное использование наследования. Сформулированный Барбарой Лисков, этот принцип подчёркивает важность взаимозаменяемости объектов.

#### Основы принципа:

Согласно LSP, **объекты производного класса** должны быть **взаимозаменяемы с объектами базового класса** без нарушения корректности работы программы. Это означает, что поведение производных классов не должно противоречить ожиданиям, основанным на поведении базовых классов.

#### Ключевые аспекты для соблюдения LSP:

- **Тщательное проектирование**: Необходимо внимательно спроектировать классы и их интерфейсы, чтобы обеспечить совместимость поведения между базовыми и производными классами.
- **Уточнение функциональности**: Производные классы должны расширять или уточнять функциональность базовых классов, не изменяя основных принципов их работы.
- **Сохранение ожидаемого поведения**: Изменения в производных классах не должны приводить к неожиданным результатам при использовании в контексте, предназначенном для базового класса.

Соблюдение принципа подстановки Лисков обеспечивает более гибкую и масштабируемую архитектуру программного обеспечения, способствуя созданию систем, которые легче поддерживать и развивать.

Чтобы проиллюстрировать принцип LSP рассмотрим две простые реализации классов (верный и неверный):

### Корректная реализация

В корректной реализации мы определим общее поведение в базовом классе Animal, которое можно корректно реализовать во всех производных классах. Например, все животные могут двигаться, но способ движения может отличаться.


````python
class Animal:
    def move(self):
        print("Это животное двигается")

class Bird(Animal):
    def move(self):
        print("Эта птица летает")

class Horse(Animal):
    def move(self):
        print("Эта лошадь бежит")

def make_animal_move(animal: Animal):
    animal.move()

# Создаем объекты
sparrow = Bird()
secretariat = Horse()

# Заставляем их двигаться
make_animal_move(sparrow)    # Вывод: Эта птица летает
make_animal_move(secretariat) # Вывод: Эта лошадь бежит
````

### Некорректная реализация
В некорректной реализации мы определим в базовом классе специфичное поведение, которое не может быть корректно реализовано во всех производных классах, например, летать.

````python
class Animal:
    def fly(self):
        print("Это животное летает")

class Bird(Animal):
    def fly(self):
        print("Эта птица летает")

class Horse(Animal):
    def fly(self):
        # Попытка реализовать метод fly для лошади, которая не может летать
        raise NotImplementedError("Лошади не умеют летать")

def make_animal_fly(animal: Animal):
    animal.fly()

# Создаем объекты
sparrow = Bird()
secretariat = Horse()

# Попытка заставить их летать
make_animal_fly(sparrow)    # Вывод: Эта птица летает
make_animal_fly(secretariat) # Возникнет исключение NotImplementedError
````

### Наследование повсюду

В экосистеме **Python**, каждый элемент — будь то класс, функция, или даже модуль — является объектом. Это уникальная особенность Python, подчёркивающая глубоко укоренённую объектно-ориентированную природу языка. Основополагающим принципом, объединяющим все эти объекты, является **наследование** — фундаментальный компонент объектно-ориентированного программирования.

#### Особенности наследования в Python:
- **Множественное наследование**: Python выделяется среди многих языков программирования поддержкой множественного наследования, позволяющего классу иметь более одного родительского класса.
- **Базовый класс `object`**: Все классы в Python, даже самые простые, неявно наследуются от класса `object`. Это делает класс `object` универсальным предком и влияет на поведение и возможности всех классов в языке.

Использование классов в Python неизбежно влечёт за собой применение наследования. Даже когда кажется, что вы создаёте класс "с нуля", без явного указания родительского класса, ваш класс наследуется от `object`, получая от него базовые атрибуты и методы.

````python
class EmptyClass:
    pass

empty_object = EmptyClass()
print(dir(empty_object))
````

Создав экземпляр `EmptyClass` и взглянув на его содержимое с помощью `dir()`, видим набор встроенных методов, которые унаследованы от `object`. Это же самое можно наблюдать и для простого создания объекта `object()` — списки методов будут практически идентичны.

````python
empty_object = object()
print(dir(empty_object))
````

````python
class EmptyClass(object):
    pass

empty_object = EmptyClass()
print(dir(empty_object))
````

Интересно, что все классы в Python по умолчанию являются наследниками `object`, и это добавляет им множество встроенных возможностей. В Python 2 для получения этих возможностей классы должны были явно наследоваться от `object`, но в Python 3 это происходит автоматически.

### Исключения - исключение в Python

В Python, где всё строится на объектно-ориентированном подходе и наследовании, исключения занимают уникальное положение. В отличие от большинства классов, которые неявно наследуются от базового класса `object`, классы исключений следуют другому правилу — они должны происходить от `BaseException` (Представьте себе, все в Python наследуется от object, а исключения - нет).

#### Особенности классов исключений:
- **Требование к иерархии**: Чтобы класс мог быть использован в качестве исключения, он должен наследоваться от `BaseException`, напрямую или через промежуточные классы, такие как `Exception`.

- **Использование `Exception` для создания собственных исключений**: На практике, для создания пользовательских классов исключений рекомендуется наследование от `Exception`, который является прямым потомком `BaseException` и предназначен для всех исключений, не связанных с завершением программы.

Пример создания собственного исключения:
```python
class MyCustomError(Exception):
    """Пользовательское исключение для демонстрации наследования от Exception."""
    pass

# Пример вызова исключения
raise MyCustomError("Произошла ошибка!")
```

## Композиция

**Композиция** — это ключевая концепция в объектно-ориентированном программировании (ООП), моделирующая отношение **"имеет"** между объектами. Это означает, что один класс (**композит**) может включать в себя один или несколько объектов других классов (**компонентов**), образуя тем самым более сложные структуры.


### Пример в UML

В **UML** диаграммах композиция представлена линией, идущей от класса-композита к классу-компоненту, с ромбиком на стороне композита. Это символизирует отношение "имеет". Кардинальность на конце линии у класса-компонента указывает на то, сколько объектов данного типа может содержать композит.

### Кардинальность отношений

Кардинальность **(количественная характеристика, указывающая на возможное количество связей между классами)** может быть представлена в нескольких формах:
- Число (**1**, **2**, **3**, ...) указывает на точное количество компонентов.
- Символ **\*** означает, что композит может содержать любое количество компонентов.
- Диапазон (**1..4**, **2..5**, **1..\***) показывает минимальное и максимальное количество компонентов.

### Вспомним про интерфейсы.

**Интерфейс класса** — это набор публичных методов и свойств, которые класс предоставляет внешнему миру. Это то, как объекты класса "общаются" с остальной частью программы. Интерфейс определяет, какие действия можно выполнять с объектами класса, но не раскрывает деталей реализации этих действий.

При работе с **наследованием**, **базовый класс определяет общий интерфейс** для всех **производных классов**. Это означает, что все классы-наследники будут иметь тот же базовый набор методов и свойств, что и их базовый класс, **но могут добавлять свои уникальные или изменять поведение унаследованных методов.**

### PayrollSystem

Давайте начнём с реализации класса `PayrollSystem`, который будет обрабатывать заработную плату:

````python
class PayrollSystem:
    def calculate_payroll(self, employees):
        print('Calculating Payroll')
        print('===================')
        for employee in employees:
            print(f'Payroll for: {employee.id} - {employee.name}')
            print(f'- Check amount: {employee.calculate_payroll()}')
            print('')
````

Здесь `PayrollSystem` — это система, отвечающая за обработку заработной платы. Метод calculate_payroll принимает список сотрудников и выводит информацию о заработной плате для каждого из них. Предполагается, что каждый сотрудник в списке будет иметь уникальный идентификатор `id`, имя `name` и метод `calculate_payroll()`, который рассчитывает заработную плату сотрудника.

Для дальнейшей работы нам потребуется определить базовый класс `Employee`, который будет представлять общий интерфейс для всех типов сотрудников. Производные классы от `Employee` могут представлять различные категории сотрудников, например, постоянных сотрудников, подрядчиков или сотрудников на частичной ставке, и предоставлять специализированные реализации метода `calculate_payroll`.

````python
class Employee:
    def __init__(self, id, name):
        self.id = id
        self.name = name
````

Этот базовый класс `Employee` определяет конструктор, который инициализирует сотрудника с заданным идентификатором и именем. **Конкретная реализация расчёта заработной платы будет зависеть от производных классов**, которые расширяют `Employee`, таким образом, демонстрируя **наследование** в объектно-ориентированном программировании.

### Проблема

В нашей HR-системе все работники, независимо от их специфики, проходят через один базовый класс Employee. Это как основа для каждого, кто работает у нас, со своим уникальным номером и именем. Это как говорить: "У каждого нашего сотрудника есть имя и свой номер".

Но не всё так просто, как кажется на первый взгляд. Мы хотим, чтобы каждый сотрудник мог показать, сколько он заработал за неделю через специальный метод `.calculate_payroll()`. И вот тут всё становится интереснее, **потому что как рассчитывается зарплата, зависит от того, кто перед вами: администратор, менеджер или кто-то ещё.**

### Решение

Допустим, у нас есть **административные сотрудники с фиксированной зарплатой**. Они каждую неделю получают одну и ту же сумму. Для них мы разработали класс `SalaryEmployee`, который является частью `Employee`. В этот класс входят базовые данные как `.id` и `.name`, плюс добавляется `weekly_salary` — их недельный заработок.



````python
class SalaryEmployee(Employee):
    def __init__(self, id, name, weekly_salary):
        super().__init__(id, name)
        self.weekly_salary = weekly_salary

    def calculate_payroll(self):
        return self.weekly_salary
````

### *Большое лирическое отступление 2:*

#### super()

С помощью `super()` мы инициализируем основные атрибуты в базовом классе `Employee`, а потом добавляем нашу особую логику расчёта заработной платы, которая в данном случае просто возвращает фиксированную сумму.

Это показывает, как мы используем наследование, чтобы сделать систему гибкой и легко масштабируемой. **Базовый класс задаёт общие правила игры**, а его **"дети" привносят что-то своё, уникальное.**

Так мы строим чёткую иерархию, где каждый сотрудник отличается по способу расчёта зарплаты, но вместе они формируют единую, хорошо организованную систему. Это даёт нам возможность без проблем добавлять новые роли в команду, не беспокоясь о том, что что-то пойдёт не так.

Рассмотрим пример классов, описывающих прямоугольник и квадрат:



````python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * self.length + 2 * self.width

class Square:
    def __init__(self, length):
        self.length = length

    def area(self):
        return self.length * self.length

    def perimeter(self):
        return 4 * self.length
````

Здесь у нас есть два похожих класса: `Rectangle` и `Square`. На первый взгляд, код кажется нормальным, но если внимательнее посмотреть, становится ясно, что квадрат — это особый случай прямоугольника, и между этими классами существует явная связь, которую мы могли бы использовать, чтобы уменьшить количество повторяющегося кода.

Используя наследование, мы можем улучшить наш код, отразив реальную взаимосвязь между прямоугольниками и квадратами:

````python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * self.length + 2 * self.width

class Square(Rectangle):
    def __init__(self, length):
        super().__init__(length, length)
````

Здесь мы используем `super()` для вызова метода `__init__()` класса `Rectangle`, позволяя нам использовать его в классе Square без повторения кода. Это показывает, как можно уменьшить количество кода, одновременно отражая связь между прямоугольниками и квадратами в реальном мире.

````python
square = Square(4)
print(square.area())
````

Мы видим, что, несмотря на то, что класс `Square` явно не реализует метод .`area()`, вызов этого метода использует метод `.area()` из **суперкласса** `Rectangle` и возвращает 16. Таким образом, класс `Square` наследует `.area()` от класса `Rectangle.`

Это прекрасный пример того, как наследование делает код более эффективным и в то же время логически отражает отношения между объектами, сокращая необходимость в дублировании кода и упрощая его поддержку и расширение.

Что же делает `super()`? Как и в других объектно-ориентированных языках есть возможность вызывать методы суперкласса в вашем подклассе. В Python это реализуется через `super()`. Основной сценарий использования — это расширение функциональности унаследованного метода.

Рассмотрим пример ниже, где вы создадите класс `Cube`, наследующий от `Square`, и расширите функциональность метода `.area()` (унаследованного от класса `Rectangle` через `Square`), чтобы рассчитать площадь поверхности и объем экземпляра `Cube`:

````python
class Square(Rectangle):
    def __init__(self, length):
        super().__init__(length, length)

class Cube(Square):
    def surface_area(self):
        face_area = super().area()
        return face_area * 6

    def volume(self):
        face_area = super().area()
        return face_area * self.length
````

````python
cube = Cube(3)
print(cube.surface_area()) # 54
print(cube.volume()) # 27
````

**Обратите внимание:** в нашем примере выше просто написать `super()` недостаточно, нужно еще указать конкретный метод из суперкласса, который мы хотим использовать

Здесь реализована два метода для класса `Cube`: `.surface_area()` и `.volume()`. Обе эти расчеты зависят от вычисления площади одной грани, поэтому вместо повторной реализации расчета площади используется `super()`, чтобы расширить вычисление площади.

Также **заметьте**, что в определении класса `Cube` нет `. __init__()`. Поскольку `Cube` наследует от `Square` и `. __init__()` на самом деле ничем не отличается от конструктора `Square`, поэтому конструктор для `Cube` можно упустить, и `. __init__()` суперкласса (`Square`) будет вызван автоматически.


#### **Погрузимся немного поглубже в super() [Deep Dive]**

**Обычно** мы используем `super()` без каких-либо дополнительных параметров, но на самом деле, `super()` **может принимать два аргумента**. **Первый** - это **класс-наследник**, а второй - это **объект этого класса**.

Давайте посмотрим на примеры, чтобы лучше понять, как это работает:




````python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * self.length + 2 * self.width

class Square(Rectangle):
    def __init__(self, length):
        super(Square, self).__init__(length, length)
````

Здесь, когда мы вызываем `super(Square, self)`, это то же самое, как если бы мы вызвали просто `super()` без параметров в Python 3. Первый параметр говорит о том, что мы работаем с классом `Square`, а второй - что мы работаем с конкретным экземпляром `Square`, то есть с `self`. И начинаем мы искать наш конструктор у родителя `Square` т.е. у `Rectangle`

````python
class Cube(Square):
    def surface_area(self):
        face_area = super(Square, self).area()
        return face_area * 6

    def volume(self):
        face_area = super(Square, self).area()
        return face_area * self.length
````

В этом случае мы меняем правила, указывая `Square` как класс для `super()`, а не `Cube`. Это означает, что `super()` начнёт искать метод `.area()` на один уровень выше `Square`, то есть в `Rectangle`.

Это может быть полезно, если, например, у `Square` есть своя версия `.area()`, от которой мы хотим отойти в `Cube`. Таким образом, мы можем контролировать, откуда берётся метод.

### Абстрактные классы

Класс `Employee` в приведенном выше примере является **абстрактным классом**. Абстрактные классы существуют для наследования, но не для создания экземпляров. Python предоставляет модуль `abc` для формального определения **абстрактных базовых классов**.

Вы можете использовать нижние подчеркивания в начале названия класса, чтобы дать понять, что объекты этого класса создавать не следует.

Подчеркивания предоставляют дружелюбный способ предотвратить неправильное использование вашего кода, но они не мешают заинтересованным пользователям создавать экземпляры этого класса.

Модуль `abc` в стандартной библиотеке Python предоставляет функциональность, чтобы предотвратить создание объектов **абстрактных классов**.

Давайте изменим реализацию класса `Employee` для корректного понимания.

<img src="https://realpython.com/cdn-cgi/image/width=960,format=auto/https://files.realpython.com/media/Duck-Typing-in-Python_Watermarked.b1b2472a1e39.jpg" width="60%" align="center" style="margin-left: 20px">

````python
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, id, name):
        self.id = id
        self.name = name

    @abstractmethod
    def calculate_payroll(self):
        pass
````

Создали класс `Employee` от `ABC`, делая его абстрактным базовым классом. Затем декорируем метод `.calculate_payroll()` декоратором `@abstractmethod`.

Это изменение имеет два приятных побочных эффекта:

1. Вы сообщаете пользователям модуля, что объекты типа `Employee` создать нельзя.
2. Вы сообщаете другим разработчикам, работающим над нашей системой, что если они наследуют от `Employee`, то они должны переопределить абстрактный метод `.calculate_payroll()`.


````python
employee = Employee(1, "Abstract")
````

Вывод показывает, что вы не можете инстанциировать класс, поскольку он содержит абстрактный метод `.calculate_payroll()`. Производные классы должны переопределить метод, чтобы позволить создание объектов их типа.

`ProductivitySystem` отслеживает производительность на основе ролей сотрудников. Существуют различные роли сотрудников:

- **Менеджеры**: Они ходят и кричат на людей, говоря им, что делать. Они являются сотрудниками с фиксированной зарплатой и зарабатывают больше денег.
- **Секретари**: Они занимаются всей бумажной работой для менеджеров и обеспечивают своевременный выставления счетов и оплату. Они также являются сотрудниками с фиксированной зарплатой, но зарабатывают меньше денег.
- **Сотрудники отдела продаж**: Они совершают множество телефонных звонков для продажи продуктов. У них есть зарплата, но они также получают комиссионные от продаж.
- **Рабочие на фабрике**: Они производят продукты для компании. Их оплата происходит почасово.

Давайте реализуем

````python
class Manager(SalaryEmployee):
    def work(self, hours):
        print(f"{self.name} screams and yells for {hours} hours.")

class Secretary(SalaryEmployee):
    def work(self, hours):
        print(f"{self.name} expends {hours} hours doing office paperwork.")

class SalesPerson(CommissionEmployee):
    def work(self, hours):
        print(f"{self.name} expends {hours} hours on the phone.")

class FactoryWorker(HourlyEmployee):
    def work(self, hours):
        print(f"{self.name} manufactures gadgets for {hours} hours.")
````

## Множественное наследование

**Множественное наследование** — это возможность класса наследоваться одновременно от нескольких базовых классов. Эта концепция позволяет одному классу объединить функциональность и свойства нескольких классов.

### Почему множественное наследование вызывает споры?

Множественное наследование часто критикуется и имеет плохую репутацию по нескольким причинам:

- **Сложность восприятия**: Когда класс наследуется от многих базовых классов, становится сложнее понимать его структуру и взаимодействие с базовыми классами.
- **Проблемы с разрешением конфликтов**: Если базовые классы имеют методы с одинаковыми именами, может возникнуть неоднозначность, какой метод будет использоваться.

### Альтернативы множественному наследованию

Большинство современных языков программирования избегают множественного наследования, предпочитая подходы, которые упрощают проектирование и повышают читаемость кода:

- **Интерфейсы**: Вместо множественного наследования, языки поддерживают реализацию нескольких интерфейсов. Интерфейс определяет набор методов, которые класс должен реализовать, но не предоставляет реализацию этих методов.
- **Композиция**: Этот подход предпочтительнее множественного наследования, поскольку он позволяет классу иметь другие классы в качестве полей, вместо того чтобы наследоваться от них, тем самым избегая сложностей множественного наследования. Мы его рассмотрим немного позже. А пока разберемся с множественным наследованием

### Порядок разрешения методов (MRO)

Порядок разрешения методов (MRO, Method Resolution Order) определяет, как Python ищет унаследованные методы. Это особенно полезно при использовании функции `super()`, потому что MRO точно указывает, где и в каком порядке Python будет искать метод, вызываемый через `super()`.

У каждого класса есть атрибут `.__mro__`, который позволяет нам проверить порядок. Например:



````python
Pyramid.__mro__
````

Это говорит нам о том, что методы будут искаться сначала в `Pyramid`, затем в `Triangle`, затем в `Square`, затем в `Rectangle`, и, если ничего не найдено, в `object`, от которого произошли все классы.

### Описание проблемы
Проблема здесь в том, что интерпретатор ищет метод `.area()` сначала в `Triangle`, а уже потом в `Square` и `Rectangle`, и, найдя `.area()` в `Triangle`, Python вызывает его вместо того, который вам нужен. Поскольку `Triangle.area()` ожидает наличия атрибутов `.height` и `.base`, Python выдаёт ошибку `AttributeError`.

К счастью, у вас есть некоторый контроль над тем, как строится MRO. Просто изменив сигнатуру класса `Pyramid`, вы можете заставить Python искать методы в нужном вам порядке, и методы будут разрешаться корректно. ### Композиция

**Композиция** — это концепция объектно-ориентированного проектирования, моделирующая отношение **"имеет"**. В композиции класс, называемый *композитным*, содержит объект или компонент другого класса. Другими словами, **композитный класс "имеет" компонент другого класса**.

Эта концепция позволяет композитным классам **повторно использовать реализацию** содержащихся в них компонентов. Композитный класс **не наследует интерфейс** компонентного класса, но может использовать его реализацию.

Связь между двумя классами в композиции считается **слабо связанной**, потому что изменения в классе компонента редко влияют на композитный класс, а изменения в композитном классе никогда не влияют на класс компонента.

Это обеспечивает **лучшую адаптивность к изменениям** и позволяет приложениям вводить новые требования, не затрагивая существующий код.

Сравнивая два конкурирующих программных решения, одно на основе **наследования**, и другое на основе **композиции**, решение на основе композиции обычно бывает **более гибким**.

Система выплат (`PayrollSystem`) теперь поддерживает внутреннюю базу данных политик оплаты труда для каждого сотрудника. Она предоставляет метод `.get_policy()`, который, получив `.id` сотрудника, возвращает его политику оплаты. Если указанный `.id` не существует в системе, то метод генерирует исключение ValueError.

Реализация .calculate_payroll() работает так же, как и раньше. Она принимает список сотрудников, рассчитывает заработную плату и печатает результаты.

````python
class PayrollPolicy:
    def __init__(self):
        self.hours_worked = 0

    def track_work(self, hours):
        self.hours_worked += hours

class SalaryPolicy(PayrollPolicy):
    def __init__(self, weekly_salary):
        super().__init__()
        self.weekly_salary = weekly_salary

    def calculate_payroll(self):
        return self.weekly_salary

class HourlyPolicy(PayrollPolicy):
    def __init__(self, hourly_rate):
        super().__init__()
        self.hourly_rate = hourly_rate

    def calculate_payroll(self):
        return self.hours_worked * self.hourly_rate

class CommissionPolicy(SalaryPolicy):
    def __init__(self, weekly_salary, commission_per_sale):
        super().__init__(weekly_salary)
        self.commission_per_sale = commission_per_sale

    @property
    def commission(self):
        sales = self.hours_worked / 5
        return sales * self.commission_per_sale

    def calculate_payroll(self):
        fixed = super().calculate_payroll()
        return fixed + self.commission
````

Сначала вы реализуете класс `PayrollPolicy`, который служит базовым классом для всех оплатытруда. Этот класс отслеживает `hours_worked`, ведь время мы трекаем у всех.

Остальные классы политик наследуются от `PayrollPolicy`. Здесь используетcя наследование, потому что хочется воспользоваться реализацией `PayrollPolicy`. Также, `SalaryPolicy`, `HourlyPolicy` и `CommissionPolicy` являются `PayrollPolicy`.

`SalaryPolicy` инициализируется значением `weekly_salary`, которое затем используется в `.calculate_payroll()`. `HourlyPolicy` инициализируется `hourly_rate` и реализует .calculate_payroll(), используя `hours_worked` базового класса.

Класс `CommissionPolicy` инициализируется параметрами `weekly_salary`, но также требует параметр `commission_per_sale`.

В примере предполагается, что продажа происходит каждые пять отработанных часов, и .`commission` является количеством продаж, умноженным на значение .`commission_per_sale`.

`CommissionPolicy` реализует метод `.calculate_payroll()`, сначала используя реализацию в `SalaryPolicy`, а затем добавляя рассчитанную комиссию.

### Что такое принцип подстановки лисков?

Принцип подстановки Лисков говорит нам, что если у нас есть классы `A` и `B`, где `B` наследуется от `A`, то мы должны иметь возможность использовать объекты `B` везде, где ожидаются объекты `A`, не нарушая работу программы. Это значит, что класс-потомок должен полностью соответствовать интерфейсу и поведению класса-родителя.



### Почему квадрат не всегда прямоугольник в программировании

На первый взгляд может показаться, что `Square` (Квадрат) является частным случаем `Rectangle` (Прямоугольника), потому что квадрат — это прямоугольник с равными сторонами.

Однако, если мы решим добавить функциональность изменения размеров для `Rectangle`, возникает проблема. Представим, что у `Rectangle` есть метод `resize()`, который позволяет изменять длину и ширину независимо друг от друга. Если мы используем этот же метод для `Square`, то сталкиваемся с тем, что после изменения размеров квадрат перестаёт быть квадратом (ведь у квадрата стороны должны быть равны).



### Как это показывает, что наследование здесь не подходит?

Тест, который предложен для проверки соответствия принципу подстановки лисков, заключается в двух вопросах:

1. **B является A?** (Квадрат является прямоугольником?)
2. **A является B?** (Прямоугoльник является квадратом?)

Если ответ на оба вопроса "да", то использовать наследование не стоит, потому что это указывает на то, что между классами существует двусторонняя зависимость, которая может привести к неправильной реализации или использованию классов.

В случае с `Rectangle` и `Square`, мы видим, что хотя квадрат и является прямоугольником в математическом смысле, в программировании это наследование вызывает проблемы при попытке изменить размеры, так как изменение размеров квадрата должно следовать другому правилу, чем изменение размеров прямоугольника.

### Как можно исправить проблему?

Вместо того чтобы использовать наследование, можно использовать композицию или более гибкие паттерны проектирования, которые позволят `Square` и `Rectangle` иметь различную функциональность изменения размеров, не нарушая общую логику программы.

### Смешивание функциональности с помощью классов Mixin

Одно из применений множественного наследования в Python - расширение функциональных возможностей классов с помощью **миксинов**. **Миксин** - это класс, который предоставляет методы другим классам, но не рассматривается как базовый класс.

**Миксин** позволяет другим классам повторно использовать его интерфейс и реализацию без превращения в суперкласс. Он реализует уникальное поведение, которое вы можете агрегировать к другим несвязанным классам. Миксины аналогичны композиции, но создают более сильную связь.

Представьте, что вы хотите преобразовать объекты определенных типов в вашем приложении в словарное представление объекта. Вы могли бы предоставить метод `.to_dict()` в каждом классе, который вы хотите поддержать этой функцией, но реализация `.to_dict()` кажется очень похожей.

Это может быть хорошим кандидатом для миксина. Вы начинаете с небольшой модификации класса `Employee` из примера композиции:

```python
class ToDictMixin:
    def to_dict(self):
        return {prop: self.__dict__[prop] for prop in self.__dict__ if not prop.startswith('_')}

class Employee(ToDictMixin):
    def __init__(self, name, id):
        self.name = name
        self.id = id

# Пример использования
employee = Employee("John Doe", 12345)
print(employee.to_dict())
```

#### Как работают классы Mixin

- **Миксины для повторного использования кода**: Миксины предлагают способ повторного использования кода между классами, которые не находятся в одной иерархии наследования.
- **Не предназначены для прямого инстанцирования**: Миксины предназначены для того, чтобы быть включенными в другие классы и не предназначены для прямого создания экземпляров.
- **Расширение функциональности**: Используя миксины, вы можете легко добавить дополнительную функциональность к существующим классам, не изменяя их иерархию наследования.

#### Преимущества использования миксинов

1. **Повторное использование кода**: Миксины позволяют избежать дублирования кода, предоставляя общую функциональность, которая может быть повторно использована в различных классах.
2. **Гибкость**: Они предоставляют гибкий способ добавления функциональности к классам без изменения их основной иерархии.
3. **Чистота дизайна**: Использование миксинов может помочь сохранить чистоту и организованность вашего кода, делая его более модульным и легким для понимания.

Миксины в Python предлагают инструмент для расширения функциональности классов, сохраняя при этом код организованным и поддерживаемым. Используя миксины, разработчики могут создавать более модульные и гибкие приложения, избегая при этом проблем, связанных с множественным наследованием и жесткой связью.

````python
class Employee:
    def __init__(self, id, name, address, role, payroll):
        self.id = id
        self.name = name
        self.address = address
        self._role = role
        self._payroll = payroll

    def work(self, hours):
        duties = self._role.perform_duties(hours)
        print(f"Employee {self.id} - {self.name}:")
        print(f"- {duties}")
        print("")
        self._payroll.track_work(hours)

    def calculate_payroll(self):
        return self._payroll.calculate_payroll()
````

Изменения минимальны. Просто изменили атрибуты .role и .payroll, сделав их внутренними, добавив ведущее подчеркивание к их именам. Чуть позже эти действия станут понятными

````python
class AsDictionaryMixin:
    def to_dict(self):
        return {
            prop: self._represent(value)
            for prop, value in self.__dict__.items()
            if not self._is_internal(prop)
        }

    def _represent(self, value):
        if isinstance(value, object):
            if hasattr(value, "to_dict"):
                return value.to_dict()
            else:
                return str(value)
        else:
            return value

    def _is_internal(self, prop):
        return prop.startswith("_")
````

### Класс `AsDictionaryMixin` и его метод `.to_dict()`

Класс `AsDictionaryMixin` предоставляет метод **`.to_dict()`**, который возвращает представление самого себя в виде словаря. Метод реализуется как генератор словаря, создающий отображение `prop` в `value` для каждого элемента в `self.__dict__.items()`, если `prop` не является внутренним.

**Примечание:** Это объясняет, почему атрибуты `role` и `payroll` внутренние в классе `Employee` — чтобы они представлялись в словаре.

### Наследование от `object`

Как видно было в начале, при создании класса он наследует некоторые методы от `object`, и одним из таких членов является **`.__dict__`**, который по сути является отображением всех атрибутов объекта в их значения.

### Итерация и фильтрация атрибутов

Вы итерируете по всем элементам в **`.__dict__`** и фильтруете те, имена которых начинаются с подчеркивания, используя метод **`._is_internal()`**.

### Метод `._represent()`

С помощью метода **`._represent()`** проверяется указанное значение. Если значение является объектом, то метод проверяет, есть ли у него член **`.to_dict()`** и использует его для представления объекта. В противном случае возвращается строковое представление. Если значение не является объектом, то просто возвращается значение.


````python
class Employee(AsDictionaryMixin):
    def __init__(self, id, name, address, role, payroll):
        self.id = id
        self.name = name
        self.address = address
        self._role = role
        self._payroll = payroll

    def work(self, hours):
        duties = self._role.perform_duties(hours)
        print(f"Employee {self.id} - {self.name}:")
        print(f"- {duties}")
        print("")
        self._payroll.track_work(hours)

    def calculate_payroll(self):
        return self._payroll.calculate_payroll()
````

````python
class Address(AsDictionaryMixin):
    def __init__(self, street, city, state, zipcode, street2=""):
        self.street = street
        self.street2 = street2
        self.city = city
        self.state = state
        self.zipcode = zipcode

    def __str__(self):
        lines = [self.street]
        if self.street2:
            lines.append(self.street2)
        lines.append(f"{self.city}, {self.state} {self.zipcode}")
        return "\n".join(lines)
````

### Наследование VS Композиция

Понимание разницы между наследованием и композицией в Python - ключевой аспект объектно-ориентированного программирования. Наследование используется для моделирования отношения "является", в то время как композиция отражает отношение "имеет".

Вот некоторые рекомендации:

- **Используйте наследование**, когда объект "является" другим объектом. Обоснуйте это отношение в обоих направлениях (вспоминаем правило подстановки Лиска)

- **Используйте наследование**, чтобы предоставить функциональность миксина нескольким несвязанным классам.

- **Используйте композицию**, чтобы моделировать отношение "имеет", когда объект содержит другой объект.

- **Используйте композицию**, чтобы создавать компоненты, повторно используемые в различных классах.

- **Используйте композицию**, чтобы реализовать группы ролей, политик зарплат и тд, применяемых к другим классам.

- **Используйте композицию**, чтобы изменять поведение во время выполнения без изменения существующих классов.
