## Что такое типизация вообще?

Типизация — это система правил в языке программирования, которая определяет, какие значения могут храниться в переменных и какие операции над ними разрешены.


### Статическая типизация


В мире программирования, как вам может быть известно мир любит делиться на ситхов и джедаев, то есть на языки с динамической типизацией и статической типизацией.

- Тип переменной известен на этапе компиляции.

- Компилятор проверяет, что все операции с переменной корректны.

- Ошибки «не тот тип» ловятся до запуска программы.

- Нужно явно указывать тип при объявлении (в большинстве языков).


Например в C#:
```csharp
int digit = 10; // digit всегда будет принимать число 10
digit = "10";  // Выдаст ошибку сборки по несовпадению типов.
```


Или в Java
```java
String name = "Anna"; // Имя является строкой
name = 42;           // и оно внезапно поменяться на число
```

Да, вы можете сказать, что в C# есть `var` который позволяет инферить тип переменной и не указывать его явно, но на самом деле, вы не можете присвоить такой переменной значение другого типа после инициализации такой переменной. Но и тут вы можете возразить мне друзья, что некоторые языки программирования умеют в `dynamic`, но что то я увлекся!

Так, что же нам дает статическая типизация:
- Ошибки находятся раньше (во время компиляции).
- IDE и компилятор могут лучше оптимизировать код.
- Код быстрее работает.
- Иногда объекты весят намного меньше

Как вы думаете в чем проблема статической типизации?
Как вы знаете лень двигатель прогресса, а когда нам нужно много писать или много что то делать, так вообще ужас!

А проблемы статической типизации следующие:
- Нужно больше кода (явные объявления типов).
- Меньше гибкости при написании быстрых прототипов.


### Динамическая типизация

**Динамическая типизация** означает:

- Тип переменной определяется в момент выполнения программы.

- Переменная может менять тип в процессе работы.

Ниже приведен пример на Python

````python
a = 12
b = "Hello world"

print("Type a is ", type(a))
print("Type b is " , type(b))

a = b
b = 13

print("Type a is ", type(a))
print("Type b is " , type(b))

print(a + " world")
# print(a + 6)
````

````python
def abc(a: int, b: int, c: int ) -> int:
  return a + b + c



abc(10, "10", 1e-10)
````

В чем же заключаются плюшки динамической типизации? Во-первых, быстрое прототипирование. Второе, не надо заморачиваться над типами переменных, интерпретатор сам разберется.

Но при этих плюсах есть также весомые недостатки:
- Ошибки типов появляются только во время выполнения
- Меньше возможностей для оптимизации → программы обычно медленнее


В мире программирования есть сильные и слабые. И если кто не понял, мы продолжаем речь про **типизацию**.

### Сильная и слабая типизация языков программирования

#### Что вообще значит «сильная/слабая»?

**Сильная типизация** — язык не делает неявных преобразований между «несовместимыми» типами. вы всегда явно преобразуете тип, иначе получите ошибку.

*Примеры*: python, java, c#, ruby.

**Cлабая типизация** — язык активно приводит типы за вас (иногда неожиданно), чтобы выражение «как-нибудь» выполнилось.
примеры: javascript (без строгих правил при ==), исторически php в обычном режиме.

````python
int("5") + 5
"ответ: " + str(42)
````

#### Базовое определение

* **Компилируемый язык** — программа **переводится целиком в машинный код** ещё до запуска, а потом запускается как самостоятельный бинарный файл.
* **Интерпретируемый язык** — программа **выполняется построчно** или блоками специальной программой (интерпретатором), без предварительного полного преобразования в машинный код.

---

#### Процесс выполнения программ

##### Компилируемый язык (C, C++, Go, Rust):

1. Вы пишете исходный код (`.c`, `.cpp`, `.rs`)
2. Компилятор переводит код в машинные инструкции (`.exe`, `.out`)
3. ОС запускает бинарник напрямую
4. Ошибки типов и синтаксиса ловятся ещё на этапе компиляции

##### Интерпретируемый язык (Python, JavaScript, Ruby):

1. Вы пишете код (`.py`, `.js`)
2. Интерпретатор читает код и выполняет его «на лету»
3. Ошибки проявляются только во время работы (runtime errors)

---

#### Сравнение

| Характеристика         | Компилируемый язык                 | Интерпретируемый язык            |
| ---------------------- | ---------------------------------- | -------------------------------- |
| Перевод в машинный код | Полный (один раз при компиляции)   | Построчно или байт-код «на лету» |
| Скорость выполнения    | Обычно выше                        | Обычно ниже                      |
| Проверка ошибок        | На этапе компиляции                | Во время работы                  |
| Удобство разработки    | Нужно компилировать перед запуском | Можно сразу запускать            |
| Примеры                | C, C++, Go, Rust, Swift            | Python, Ruby, JavaScript         |

---

#### Но всё сложнее: JIT и байт-код

Современные языки часто сочетают подходы:

* **Java, C#**: исходный код → компилятор → байт-код → виртуальная машина (JVM, CLR) → JIT-компиляция в машинный код.
* **Python**: исходный код → интерпретатор CPython → байт-код `.pyc` → интерпретация (есть проекты JIT — PyPy).
* **JavaScript** (Chrome V8): исходный код → парсинг → JIT-компиляция в машинный код → кеширование.

---



#### Python как пример интерпретируемого языка

1. Пишем файл `hello.py`:

```python
print("Hello, World!")
```

2. Запускаем `python hello.py`

3. Под капотом (CPython):

   * Исходный код → лексер → токены
   * Парсер → дерево (AST)
   * Компилятор CPython → байт-код (`.pyc`)
   * Виртуальная машина (Python Virtual Machine, PVM) исполняет байт-код

Поэтому Python формально **не чисто интерпретируемый**, а скорее **байт-кодовый язык с интерпретатором**.

---


#### Примеры на практике

##### Компиляция (C++)

```cpp
#include <iostream>
using namespace std;

int main() {
    cout << "Hello, World!" << endl;
    return 0;

```

Компилируем и запускаем:

```bash

```

##### Интерпретация (Python)

```python
print("Hello, World!")
```

Запуск без компиляции:

```bash
python hello.py
```

---



#### Промежуточные варианты

* **Java**: компилируется в байт-код (`.class`), исполняется JVM (с JIT).
* **C#**: компилируется в IL (Intermediate Language), исполняется CLR (с JIT).
* **JavaScript**: сначала интерпретация, но V8 и другие движки делают JIT-компиляцию.
* **Go**: всегда компилируется в машинный код (как C).

## Фундамент языка

### Переменные

В Python переменные используются для хранения данных в памяти компьютера, предоставляя способ помечать и ссылаться на эти данные в программе. Можно представить переменные как подписанные контейнеры, которые содержат значения, доступные для использования и изменения по мере необходимости.

Переменные позволяют временно хранить информацию, выполнять с ней операции и обращаться к ней позже. Эта фундаментальная концепция является ключевой для написания динамичных программ, которые могут обрабатывать и преобразовывать данные.

Когда вы создаёте переменную в Python, вы фактически резервируете место в памяти компьютера и присваиваете этому месту имя. Python использует динамическую типизацию, что означает, что вам не нужно явно указывать тип переменной — Python определяет его исходя из присвоенного значения.


### Важные компоненты

* **Имена переменных:** В Python имена переменных:

  * Могут содержать буквы, цифры и символ подчёркивания (`a-z`, `A-Z`, `0-9`, `_`)
  * Должны начинаться с буквы или подчёркивания (но не с цифры)
  * Чувствительны к регистру (`myVar` и `myvar` — это *разные* переменные)
  * Не могут совпадать с зарезервированными словами Python (например, `if`, `for`, `class`)
  * По соглашению используются в стиле `snake_case` (строчные буквы с подчёркиваниями)

* **Оператор присваивания:** Символ `=` используется для присвоения значений переменным

  * Пример: `x = 10` присваивает переменной `x` значение `10`

* **Управление памятью:** Python автоматически управляет выделением памяти

  * Не нужно объявлять переменные заранее
  * Не нужно указывать тип переменной
  * Память освобождается автоматически, когда переменные больше не используются

| Особенность                | Описание                                        | Пример                                         |
| -------------------------- | ----------------------------------------------- | ---------------------------------------------- |
| Именование                 | Правила для корректных имён переменных          | `student_count` (верно), `1st_place` (неверно) |
| Присваивание               | Как значения присваиваются переменным           | `x = 42`                                       |
| Переприсваивание           | Значения переменных можно менять                | `x = 10`, затем `x = "hello"`                  |
| Множественное присваивание | Присваивание нескольким переменным одновременно | `x, y, z = 1, 2, 3`                            |

#### Создание переменных в Python

В Python переменная создаётся путём присвоения ей значения с помощью оператора `=`:


Здесь:

Обратите внимание: вам не нужно заранее объявлять переменные или указывать их тип — Python определяет его автоматически на основе присвоенного значения.

````python
# Creating different variables
age = 25               # Integer
height = 1.75          # Float
name = "Alice"         # String
is_student = True      # Boolean
````

Обратите внимание, что нам не нужно указывать тип переменной — Python определяет его автоматически на основе присвоенного значения. Это делает код на Python более кратким и гибким по сравнению со статически типизированными языками.

#### Частые ошибки при использовании имён переменных

* **Использование зарезервированных ключевых слов**: Никогда не используйте ключевые слова Python, такие как `if`, `for`, `class`, в качестве имён переменных
* **Начало имени с числа**: Имена переменных не могут начинаться с цифры (например, `1variable` — недопустимо)
* **Использование пробелов или спецсимволов**: Вместо пробелов используйте подчёркивания, избегайте специальных символов
* **Слишком абстрактные имена переменных**: Имена вроде `x`, `y`, `z` затрудняют понимание кода

Преобразование типов

Иногда возникает необходимость преобразовать данные из одного типа в другой. Для этого в Python предусмотрены встроенные функции:

````python
# Из числа в строку
x = 10
print(str(x))   # "10"

# Из строки в число
y = "25"
print(int(y))   # 25

# Из числа с плавающей точкой в целое (округление вниз)
z = 3.99
print(int(z))   # 3

# Из числа в float
a = 7
print(float(a))  # 7.0

# Из списка в множество (удаляются дубликаты)
lst = [1, 2, 2, 3]
print(set(lst))  # {1, 2, 3}

# Из множества в список
s = {10, 20, 30}
print(list(s))   # [10, 20, 30] (порядок может отличаться)

# Из кортежа в список
t = (1, 2, 3)
print(list(t))   # [1, 2, 3]
````

### Арифметические операторы

Операторы в Python — это специальные символы, которые выполняют действия над переменными и значениями. Понимание операторов необходимо для выполнения вычислений, сравнений и управления логикой программ.

Арифметические операторы используются для выполнения математических операций, таких как сложение, вычитание, умножение, деление и другие. Эти операторы составляют основу численных вычислений в Python.

---

### Важные компоненты

Вот полный список арифметических операторов в Python:

| Оператор | Название              | Пример   | Результат   | Описание                                                                 |
| -------- | --------------------- | -------- | ----------- | ------------------------------------------------------------------------ |
| `+`      | Сложение              | `5 + 3`  | `8`         | Складывает два операнда                                                  |
| `-`      | Вычитание             | `5 - 3`  | `2`         | Вычитает правый операнд из левого                                        |
| `*`      | Умножение             | `5 * 3`  | `15`        | Умножает два операнда                                                    |
| `/`      | Деление               | `5 / 3`  | `1.6666...` | Делит левый операнд на правый                                            |
| `//`     | Целочисленное деление | `5 // 3` | `1`         | Делит и возвращает наибольшее целое число, меньшее или равное результату |
| `%`      | Остаток (modulus)     | `5 % 3`  | `2`         | Возвращает остаток от деления                                            |
| `**`     | Возведение в степень  | `5 ** 3` | `125`       | Возводит левый операнд в степень правого                                 |

---

* **Операторы сравнения**: используются для сравнения значений и возвращают логические результаты (`True` или `False`)
* **Логические операторы**: применяются для объединения условных выражений
* **Операторы присваивания**: применяются для присвоения значений переменным
* **Приоритет операций**: определяет порядок выполнения действий

### Основное содержание

#### Арифметические операторы

Посмотрим на примеры их использования:





````python
a = 10
b = 3

print(a + b)   # 13 (сложение)
print(a - b)   # 7  (вычитание)
print(a * b)   # 30 (умножение)
print(a / b)   # 3.333... (деление с плавающей точкой)
print(a // b)  # 3 (целочисленное деление)
print(a % b)   # 1 (остаток от деления)
print(a ** b)  # 1000 (10 в степени 3)
````

#### Приоритет операторов

Когда выражение содержит несколько операторов, порядок вычислений определяется правилами приоритета. Точно так же, как в математике, некоторые операции выполняются раньше других.

Вот упрощённый список приоритетов от самого высокого к самому низкому:

1. **Скобки:** `()`
2. **Возведение в степень:** `**`
3. **Унарные плюс и минус:** `+x`, `-x`
4. **Умножение, деление, целочисленное деление, остаток:** `*`, `/`, `//`, `%`
5. **Сложение и вычитание:** `+`, `-`
6. **Операторы сравнения:** `==`, `!=`, `>`, `<`, `>=`, `<=`
7. **Логическое отрицание:** `not`
8. **Логическое И:** `and`
9. **Логическое ИЛИ:** `or`

---


````python
print(2 + 3 * 4)     # 14 (сначала умножение, потом сложение)
print((2 + 3) * 4)   # 20 (скобки меняют порядок)
print(2 ** 3 ** 2)   # 512 (степени вычисляются справа налево: 3**2 = 9, потом 2**9)
print(not (5 > 3))   # False (сначала сравнение, потом not)
print(True or False and False)  # True (сначала and, потом or)
````

Перед тем как перейти к условным операторам, важно понять, что такое **логические выражения (Boolean expressions)**. Это выражения, которые вычисляются либо в `True`, либо в `False`, и они являются основой для принятия решений в программе.

Логические выражения обычно используют **операторы сравнения**, которые мы изучили ранее:

* `==` (равно)
* `!=` (не равно)
* `<` (меньше чем)
* `>` (больше чем)
* `<=` (меньше или равно)
* `>=` (больше или равно)

Также они могут включать **логические операторы** для объединения условий:

* `and` (оба условия должны быть истинными)
* `or` (достаточно, чтобы хотя бы одно условие было истинным)
* `not` (отрицание условия)

Такие выражения позволяют программе проверять условия и принимать решения на основе того, выполняются они или нет.

### 🔍 Оператор `if`

`if` — это самый простой способ условного выполнения кода в Python. Он позволяет выполнить блок кода только в том случае, если условие истинно.

**Синтаксис:**

```python
if условие:
    # блок кода, который выполнится, если условие == True
```



````python
age = 18

if age >= 18:
    print("Доступ разрешён")
````

### Оператор `if...else`

Иногда нужно выполнить один блок кода, если условие истинно, и другой — если условие ложно. Для этого используется конструкция `if...else`.

**Синтаксис:**

```python
if условие:
    # код, если условие == True
else:
    # код, если условие == False
```


---

### Оператор `if...elif...else`

Когда нужно проверить несколько условий подряд, используется `elif` (сокращение от *else if*).

**Синтаксис:**

```python
if условие1:
    # код, если условие1 == True
elif условие2:
    # код, если условие2 == True
elif условие3:
    # код, если условие3 == True
else:
    # код, если все условия == False
```


````python
age = 16

if age >= 18:
    print("Доступ разрешён")
else:
    print("Доступ запрещён")
````

````python
score = 75

if score >= 90:
    print("Отлично")
elif score >= 70:
    print("Хорошо")
elif score >= 50:
    print("Удовлетворительно")
else:
    print("Неудовлетворительно")
````

### Вложенные условия

Условные операторы можно помещать внутрь других условных операторов. Это называется **вложенностью** и позволяет строить более сложную логику принятия решений.

Пример:

````python
x = 10
y = 5

if x > 5:
    print("x больше 5")
    if y > 5:
        print("y тоже больше 5")
    else:
        print("y не больше 5")
else:
    print("x не больше 5")
````



В этом примере:

1. Сначала проверяется условие `x > 5`. Так как это `True`, выводится `"x больше 5"`.
2. Затем выполняется вложенное условие: проверяется `y > 5`.

   * Так как `y = 5`, условие ложно, и выводится `"y не больше 5"`.

---

### Важный момент

Вложенные условия позволяют строить **иерархические структуры решений**, где проверка некоторых условий происходит только при выполнении других.

Однако, слишком глубокая вложенность делает код трудным для чтения и понимания, поэтому использовать её стоит осторожно.




###  Тернарный оператор (сокращённая форма `if...else`)

В Python есть удобная конструкция для записи простых условий в одну строку. Она называется **тернарный оператор** или **условное выражение**.

**Синтаксис:**

```python
значение_if_true if условие else значение_if_false
```


````python
age = 18

# обычный if...else
if age >= 18:
    status = "Взрослый"
else:
    status = "Несовершеннолетний"

print(status)  # Взрослый
````


Теперь то же самое через тернарный оператор:

````python
age = 18
status = "Взрослый" if age >= 18 else "Несовершеннолетний"
print(status)  # Взрослый
````

### Циклы в Python

Циклы — это мощные управляющие конструкции, которые позволяют выполнять блок кода многократно. Они незаменимы для автоматизации повторяющихся задач, обработки коллекций данных и эффективной реализации алгоритмов.

---

### Цикл `for`

Цикл **`for`** используется для перебора элементов последовательности (списка, кортежа, словаря, множества или строки), а также любых других **итерируемых объектов**.
Он выполняет блок кода для **каждого элемента** последовательности.

**Синтаксис:**

```python
for item in sequence:
    # код, который выполняется для каждого элемента
```


````python
# Перебор списка
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# Перебор строки
for char in "Python":
    print(char)

# Перебор диапазона чисел
for i in range(5):
    print(i)   # вывод: 0, 1, 2, 3, 4
````

Цикл `for` часто используется вместе с функцией `range()`, которая генерирует последовательность чисел.



### Функция `range()`

Функция **`range()`** создаёт последовательность чисел, которая часто используется в цикле `for`.
У неё есть три варианта использования:

1. `range(stop)` → числа от `0` до `stop - 1`
2. `range(start, stop)` → числа от `start` до `stop - 1`
3. `range(start, stop, step)` → числа от `start` до `stop - 1` с шагом `step`



````python
# 1. От 0 до 4
for i in range(5):
    print(i)   # 0, 1, 2, 3, 4

# 2. От 2 до 6
for i in range(2, 7):
    print(i)   # 2, 3, 4, 5, 6

# 3. От 1 до 9 с шагом 2
for i in range(1, 10, 2):
    print(i)   # 1, 3, 5, 7, 9

# 4. Обратный счёт
for i in range(10, 0, -2):
    print(i)   # 10, 8, 6, 4, 2
````

Обычно `range()` применяется, когда нужно пройтись по индексам списка или выполнить цикл фиксированное количество раз.
Пример с использованием индексов:



````python
fruits = ["apple", "banana", "cherry"]

for i in range(len(fruits)):
    print(i, fruits[i])
````

### Цикл `while`

Цикл **`while`** выполняет блок кода до тех пор, пока условие остаётся истинным (`True`).

**Синтаксис:**

```python
while условие:
    # код, который выполняется, пока условие истинно
```

````python
count = 0

while count < 5:
    print("Счётчик:", count)
    count += 1
````

### Разница между `for` и `while`

* **`for`** используется, когда количество повторений известно заранее или когда нужно пройти по коллекции (список, строка и т.д.).
* **`while`** используется, когда нужно повторять действия, пока выполняется условие (количество повторений заранее может быть неизвестно).

````python
for i in range(5):
    print("for:", i)

# while — пока условие истинно
i = 0
while i < 5:
    print("while:", i)
    i += 1
````

###  Оператор `break`

Завершает цикл досрочно, даже если условие ещё истинно или элементы не закончились.


````python
for i in range(10):
    if i == 5:
        break
    print(i)
````


### Оператор `continue`

Пропускает текущую итерацию и переходит к следующей.

````python
for i in range(5):
    if i == 2:
        continue
    print(i)
````


###  Циклы с `else`

У циклов `for` и `while` в Python можно использовать блок `else`.
Он выполняется только если цикл завершился **естественным образом** (без `break`).

````python
for i in range(5):
    print(i)
else:
    print("Цикл завершён")
````

````python
Но если был `break`, блок `else` не выполнится:
````

````python
for i in range(5):
    if i == 3:
        break
    print(i)
else:
    print("Цикл завершён")
````
